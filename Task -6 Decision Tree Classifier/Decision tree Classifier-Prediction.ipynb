<p>This task is to <b>build a Decision Tree Classifier and to work on Iris dataset</b>.</p>

<h4>Let's import the required libraries.</h4>

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Iris.csv")
```

<p>We can see that data is successfully read and let's see the data</p>

```python
df.head()
```

<h4>Let's import DecisionTreeClassifier from sklearn and train the model on the available data.</h4>

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
features=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
x=df[features]
y=df.Species
dt.fit(x,y)
```

<p>Let's give it a try and see whether it can classify </p>

```python
te=np.array([7.0,3.2,4.7,1.4]).reshape(1,-1)
dt.predict(te)
```

<h4>That's done successfully.Let's visualize the tree</h4>
<p>I have used <b>tree</b> module and <b>matplotlib</b> library.</p>

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(25,15))
tree.plot_tree(dt,feature_names=features,class_names=y.unique(),rounded=True,filled=True,fontsize=15)
plt.show()
```
